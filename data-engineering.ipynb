# Data Cleaning

## 1) Quality

#### Starting the clean dataset by removing unnecceasary columns.

```python
import numpy as np
np.set_printoptions(threshold=np.inf)
```

```python
# Clean a copy of the dataset to maintain the original one.
# Removing unnessecary columns from the dataset.
movies_clean = movies.drop(['director_facebook_likes', 'actor_3_facebook_likes', 'actor_2_name', 'actor_1_facebook_likes', 'num_user_for_reviews', 'actor_2_facebook_likes', 'cast_total_facebook_likes', 'num_voted_users', 'actor_3_name', 'movie_imdb_link', 'plot_keywords', 'facenumber_in_poster', 'movie_facebook_likes', "num_critic_for_reviews", "aspect_ratio"], axis = 1)
movies_clean.shape
```

#### Then remove rows that have 4 or more Null values.

```python
s = movies_clean.isnull().sum(axis=1) > 4
print(s[s].shape)
# There are rows that have 4 or more missing features from 15 features.
```

```python
# Drop all the rows that have 4 or more missing values ([4, 279, 2241, 4945]).
movies_clean = movies_clean[movies_clean.isnull().sum(axis=1) <= 4]
```

```python
movies_clean.shape
```

```python
movies_clean = movies_clean.reset_index()
movies_clean.head(2)
```

```python
movies_clean = movies_clean.drop(['index'], axis = 1)
movies_clean.head(2)
```

```python
movies_clean['country'].unique()
```

#### Aruba is an island belong to Netherlands.
#### Hong kong is in China ofcourse.
#### Put Official Site and New Line as null values as we can handle them later.

```python
movies_clean['country'] = movies_clean.country.replace('Aruba','Netherlands')
movies_clean['country'] = movies_clean.country.replace('Official site',np.nan)
movies_clean['country'] = movies_clean.country.replace('Hong Kong','China')
movies_clean['country'] = movies_clean.country.replace('New Line', np.nan)
```

#### Full country names sometimes, abbreviations other times.

```python
movies_clean['country'] = movies_clean.country.replace('USA','United State of America')
movies_clean['country'] = movies_clean.country.replace('UK','United Kingdom')
```

```python
movies_clean['country'].unique()
```

```python
movies_clean['language'].unique()
# maya -> mayan, None -> np.nan, Panjabi -> Punjabi
```

#### There are langauges are uncorrectly written like "maya" and "panjabi", and there is "None" in the values.

```python
movies_clean['language'] = movies_clean.language.replace('maya','mayan')
movies_clean['language'] = movies_clean.language.replace('Panjabi','Punjabi')
```

# Classification.
### classification is a bouns phase in this project. The client was only need a cleaned dataset.

```python
import xgboost as xgb
import numpy as np
from scipy.stats import uniform, randint
from sklearn.metrics import auc, accuracy_score, confusion_matrix, mean_squared_error
from sklearn.model_selection import cross_val_score, GridSearchCV, KFold, RandomizedSearchCV, train_test_split
```

```python
conditions = [
    (movies_clean['imdb_score'] >= 7.8),
    ((movies_clean['imdb_score'] < 7.8) & (movies_clean['imdb_score'] >= 4.5)),
    (movies_clean['imdb_score'] < 4.5)]

# create a list of the values we want to assign for each condition
values = [0, 1, 2]

# create a new column and use np.select to assign values to it using our lists as arguments
movies_clean['IMDB_Category'] = np.select(conditions, values)

# display updated DataFrame
movies_clean.head(3)
```

```python
train, test= train_test_split(movies_clean, test_size=0.3, random_state=1)
```

```python
train.head()
```

```python
X_train = train[['director','duration','gross','main actor','language','country','content_rating','budget','title_year','genre']]
y_train = train['IMDB_Category']
X_test = test[['director','duration','gross','main actor','language','country','content_rating','budget','title_year','genre']]
y_test = test['IMDB_Category']
```

```python
# from sklearn.utils import class_weight


# class_weights = class_weight.compute_class_weight('balanced',
#                                                  classes = train.IMDB_Category.unique().tolist(),
#                                                  y = train.IMDB_Category.tolist())

# weights = np.ones(y_train.shape[0], dtype = 'float')
# for i, val in enumerate(y_train):
#     weights[i] = class_weights[val-1]
# print(class_weights)
```

```python
from sklearn.utils import class_weight
classes_weights = class_weight.compute_sample_weight(
    class_weight='balanced',
    y=train['IMDB_Category']
)
print(class_weights)
```

```python
# Create Decision Tree classifer object
xgb_model = xgb.XGBClassifier(objective="multi:softprob", random_state=42)
xgb_model.fit(X_train, y_train, sample_weight = classes_weights)

y_pred = xgb_model.predict(X_test)

print(confusion_matrix(y_test, y_pred))
```

```python
print("Accuracy:",accuracy_score(y_test, y_pred))
```
